```python
from ultralytics import YOLO
import pandas as pd
import json
import os


dataset_path = os.path.abspath('.')
```

```python
model = YOLO('yolov8n-cls.pt')

results = model.train(data=dataset_path, epochs=20, imgsz=128, batch=32)
```

```python

csv_path = 'runs/classify/train/results.csv'
df = pd.read_csv(csv_path)

df.columns = df.columns.str.strip()

yolo_log = []
for index, row in df.iterrows():
    yolo_log.append({
        "epoch": int(row['epoch']),
        "train_acc": 1.0 - float(row['train/loss']),
        "test_acc": float(row['metrics/accuracy_top1'])
    })

with open("yolo_metrics.json", "w") as f:
    json.dump(yolo_log, f, indent=2)

print("YOLO JSON saved successfully!")
```

```python
metrics = model.val()
accuracy = metrics.top1
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
